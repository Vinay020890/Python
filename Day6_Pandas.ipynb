```python
titanic['sex'].value_counts()
```

```python
titanic.groupby("sex")['survived'].mean()*100

#() - categorical columns
#[] - numerical columns
```

```python
titanic.groupby(["sex", "pclass"])['survived'].mean()*100

#() - categorical columns
#[] - numerical columns
```

```python
titanic.groupby(["pclass", "sex"])['survived'].mean()*100

#() - categorical columns
#[] - numerical columns
```

```python
titanic.groupby(["pclass", "sex"])['survived'].mean().unstack()

#() - categorical columns
#[] - numerical columns
```

```python
age_bins = [0,12,18,35,60,80]
age_labels = ["child", "Teen", "Young", "Adult", "Senior"]

titanic["age_group"] = pd.cut(titanic["age"],
                              bins=age_bins,
                              labels= age_labels)
```
